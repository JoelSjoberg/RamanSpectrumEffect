```python
from Scripts.essentials import *
from Scripts.vars import *
from sklearn.metrics import balanced_accuracy_score
from sklearn.model_selection import StratifiedKFold
p = "Data/"
title = "MANUAL"

X_all = np.concatenate([np.load(p + "train_x_"+title+".npy"), np.load(p + "val_x_"+title+".npy"), np.load(p + "test_x_"+title+".npy")])
lgm_all = np.concatenate([np.load(p + "train_lgm.npy"), np.load(p + "val_lgm.npy"), np.load(p + "test_lgm.npy")])
y_all = np.concatenate([np.load(p + "train_y_46.npy"), np.load(p + "val_y_46.npy"), np.load(p + "test_y_46.npy")])
lgm_all = np.argmax(lgm_all, axis = -1)
y_all = np.argmax(y_all, axis = -1)
lgm_all = np.where(lgm_all > 2, 0, 1)

x_axis = pd.read_excel("Data/RamanFreqTable.xlsx")["Wave"].values
```

```python
from sklearn.tree import DecisionTreeClassifier

from joblib import dump, load

# Get the folds
organizer = StratifiedKFold(n_splits=n_folds, shuffle = True, random_state = 42)
folds = organizer.split(X_all, y_all)

plt.figure(figsize = (8, 8))
for en, fold in enumerate(folds):

    train_ix = fold[0]
    val_ix = fold[1]

    # Training set
    train_x = X_all[train_ix]
    train_y = y_all[train_ix]
    train_lgm = lgm_all[train_ix]

    # Validation set
    val_x = X_all[val_ix]
    val_y = y_all[val_ix]
    val_lgm = lgm_all[val_ix]

    # Data shuffle
    ix = np.arange(len(train_x))
    np.random.shuffle(ix)
    train_x = train_x[ix]
    train_y = train_y[ix]
    train_lgm = train_lgm[ix]


    # Compute the weights for each category type
    counts = np.bincount(train_y)
    id_max_ratio = np.max(1/(counts/np.max(counts)))
    class_weights = np.sqrt((1/(counts/np.max(counts))))
    cw_id = {}
    for i in range(len(class_weights)):
        cw_id[i] = class_weights[i]
        
    counts = np.bincount(train_lgm)
    class_weights = np.sqrt((1/(counts/np.max(counts))))
    cw_lgm = {}
    for i in range(len(class_weights)):
        cw_lgm[i] = class_weights[i]

    
    accuracies = []
    accuracies_lgm = []
    for i in range(1738):
        d = train_x[:, i]
    
        if np.sum(d) == 0:
            accuracies.append(0)
            accuracies_lgm.append(0)
            continue # Do not learn 0-features
    
        # Normalize the features and save the min and max to preprocess val and test data
        min_ = np.min(d)
        max_ = np.max(d)
        d = (d - min_)/(max_ - min_)
        
        model = DecisionTreeClassifier(random_state=0, class_weight=cw_id)
        
        model.fit(d.reshape((-1, 1)), train_y)

        # Evaluate on the test data
        preds = model.predict(((val_x[:, i] - min_) / (max_ - min_)).reshape((-1, 1)))
        
        acc = balanced_accuracy_score(val_y, preds)
        print(i, ":", acc)
        accuracies.append(acc)


        model = DecisionTreeClassifier(random_state=0, class_weight=cw_lgm)
        model.fit(d.reshape((-1, 1)), train_lgm)
        # Takes a lot of memory, do not use
        #dump(model, "Models/OneFeature/TreeOneFeature_LGM_"+str(i)+"_fold_"+str(en)+".joblib")
        preds = model.predict(((val_x[:, i] - min_) / (max_ - min_)).reshape((-1, 1))) 
        acc_lgm = balanced_accuracy_score(val_lgm, preds)
        print(i, ":", acc_lgm)
        accuracies_lgm.append(acc_lgm)

    plt.plot(accuracies)
    plt.show()

    plt.plot(accuracies_lgm)
    plt.show()

    np.save("Results/(MANUAL)SingleFeatureMetrics/single_id_fold_"+str(en)+".npy", accuracies)
    np.save("Results/(MANUAL)SingleFeatureMetrics/single_lgm_fold_"+str(en)+".npy", accuracies_lgm)
    
```
